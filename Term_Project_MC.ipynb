```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

inflation = pd.read_csv("https://ai-project-inflation.s3.amazonaws.com/data_merged.csv", index_col=0)
inflation.head()
```

```python
inflation.tail()
```

```python
inflation.info()
```

```python
inflation.hist(bins=50, figsize=(20,15))
#save_fig("attribute_histogram_plots")
plt.show()
```

```python
corr_matrix = inflation.corr()
corr_matrix["Inflation_Rate"].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix

attributes = ["Inflation_Rate","10_year_bond", "Prime_Rate", "CAD_M1_MoneySupply",
              "Crude_prices"]

scatter_matrix(inflation[attributes], figsize=(12, 8))
#save_fig("scatter_matrix_plot")
```

```python
inflation.plot(kind="scatter", y="Inflation_Rate", x="10_year_bond",
             alpha=0.5)
plt.axis([0, 8, 0, 8])
```

```python
inflation.plot(kind="scatter", y="Inflation_Rate", x="Crude_prices",
             alpha=0.5)
plt.axis([0, 100, 0, 8])
```
